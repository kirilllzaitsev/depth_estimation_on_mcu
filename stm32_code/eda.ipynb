```python
import numpy as np
import matplotlib.pyplot as plt
import os
from pathlib import Path
from glob import glob
import cv2
import torch
from PIL import Image
from torchvision import transforms
```

```python
classes = [
            "T-shirt/top",
            "Trouser",
            "Pullover",
            "Dress",
            "Coat",
            "Sandal",
            "Shirt",
            "Sneaker",
            "Bag",
            "Ankle boot",
        ]
idx2class = {i: classes[i] for i in range(len(classes))}
```

```python
x_test=np.load("x_test_fmnist.npy")
y_test=np.load("y_test_fmnist.npy").squeeze()
x_test.shape, y_test.shape
```

```python
n_samples = 10
n_samples_per_row = 5
fig, axs = plt.subplots(n_samples // n_samples_per_row, n_samples_per_row, figsize=(15, 15))
for i, ax in enumerate(axs.flatten()):
    ax.imshow(x_test[i].squeeze(), cmap="gray")
    ax.set_title(f"Label: {idx2class[y_test[i]]}")
    ax.axis("off")
plt.tight_layout()
plt.show()
```
